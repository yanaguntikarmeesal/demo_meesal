Q 3)

* a toy vendor supplies three types of toys
* battery based toys keybased toys and
* electrical charging based toys the vendor gives a
* discount of 10% on order for
* battery based toys if the order is for more than rs 1000
* on orders for more than rs
* 100 for keybased toys a discount of 5% is given
* and a discount of 10% is given on
* orders for electrical charging based toys of value more
* than rs 500 assume that the
* numeric codes 1,2, and 3 are used for battery
* based toys keybased toys and
* electrical charging based toys respectively
* write a program that reads the products
* code and the order amount and prints out
* the net amount that the customer is
* required to pay after the discount




```python
product_code = int(input("Enter the product code \n1 for battery-based, \n2 for key-based, \n3 for electrical charging-based \n choice the number : "))

order_amount = float(input("Enter the order amount: "))

net_amount = order_amount

if product_code == 1: # Battery-based toys
    if order_amount > 1000:
        net_amount = order_amount * 0.90  # 10% discount (i.e., pay 90%)

elif product_code == 2: # Key-based toys
    if order_amount > 100:
        net_amount = order_amount * 0.95  # 5% discount (i.e., pay 95%)

elif product_code == 3: # Electrical charging-based toys
    if order_amount > 500:
        net_amount = order_amount * 0.90  # 10% discount (i.e., pay 90%)

else:
    print("Invalid product code entered.")

#print(f"The net amount to pay after discount is: Rs {net_amount:.2f}")
print("total amount",net_amount)
```

Q 4) the paschim gujarat vij company ltd computes the electricity bill based on the following matrix ?

```python
 # Q 4) the paschim gujarat vij company ltd computes the electricity bill based on the following matrix ?

# until consumed charges
# 0 to 100 = 0.5 per until
# 101 to 200 = rs50 plus rs:1 per until over 100 until
# 201 to 300 = rs:150 plus rs:1.50 per until over until
# 300 rs: 300 plus rs 2 per until over 300 until

# ask user to enter the past meter reading and current meter reading
# find the until consumed
# computer the bill according to matrix

```

```python
# step 1: enter the past & current meter reading
past_reading = int(input("Enter the past meter reading: "))
current_reading = int(input("Enter the current meter reading: "))

# step 2: calculate untils consumed
units_consumed = current_reading - past_reading

# step 3: computer the bill according to the given matix

if units_consumed <= 100:
    bill = units_consumed * 0.50

elif units_consumed <= 200:
    bill = 50 + (units_consumed) * 1.00

elif units_consumed <= 300:
    bill = 150 + (units_consumed ) * 1.50

else:
    bill = 300 + (units_consumed) * 2.00

# step 4: print the bill
print("Units Consumed: ",units_consumed)
print("Electricity Bill Amount: Rs ",bill)
```

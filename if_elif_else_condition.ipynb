Q 1)  enter two value from the user and if the value will be greater print greater if the value is not greater print no greater and if the values are equal print equal ?

```python
# Q 1) enter two value from the user and if the value will be greater print greater if the value is not greater print no greater and if the values are equal print equal ?

a = int(input("enter the first value : "))
b = int(input("enter the second value : "))
if a > b:
    print("the value a is greater")
elif a == b:
    print("the value is equal")
else:
    print("value is less ")
```

Q 2) take a value from the user and check it is divisible by 2 or 3 or neither 2 or 3 ?

```python
# Q 2) take a value from the user and check it is divisible by 2 or 3 or neither 2 or 3 ?

a = int(input("enter the value : "))
if a % 2 == 0:
    print("the value is divisible by 2")
elif a % 3 == 0:
    print("the value is divisible by 3")
else:
  print("the value is not divisible by 2 or 3")
```

## Q 3) enter the marks of the student if marks is greater than or 85 print a grade, marks is greater than or equal to 80 print b grade and so on till el grade ?

```python
# Q 3) enter the marks of the student if marks is greater than or 85 print a grade, marks is greater than or equal to 80 print b grade and so on till el grade ?

marks = int(input("enter the marks : "))
if marks >= 85:
    print("a grade")
elif marks >= 80 and marks <85:
    print("b grade")
elif marks >= 75 and marks < 80:
    print("c grade")
elif marks >= 70 and marks < 75:
    print("d grade")
elif marks >= 65 and marks < 70:
    print("e grade")
else:
    print("fail")
```

Q 4) write a program and take the number from the user in between 1 to 7 if number will be 1 print monday if number will be 2 print tuesday and so on ?

```python
# Q 4) write a program and take the number from the user in between 1 to 7 if number will be 1 print monday if number will be 2 print tuesday and so on ?

a = int(input("enter the number : "))
if a == 1:
    print("monday")
elif a == 2:
    print("tuesday")
elif a == 3:
    print("wednesday")
elif a == 4:
    print("thursday")
elif a == 5:
    print("friday")
elif a == 6:
    print("saturday")
elif a == 7:
    print("sunday")
else:
    print("no day")
```

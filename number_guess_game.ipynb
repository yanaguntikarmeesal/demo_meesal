Q) you are creating a  number guessing game where the computer generates a random number between 1 and 100 and the player has to guess it
Write python code to generate a random number between 1 to 100 use control flow statement and loops to allow the player to guess the number and provide feedback too hight too low or correct define a user define function to encapsulate the game logic  and call this function to play the games.


```python
import random

def number_guess_game():
    secret_number = random.randint(1,100)
    print("the secret number is = ",secret_number)

    print("welcome to the number guessing game")
    print("guess number between 1 to 100 ")

    attempts = 0
    guess = None

    while guess != secret_number:
        guess = int(input("enter your number : "))
        print("your number is : ",guess)
        attempts = attempts + 1

        if guess < secret_number:
            print("too low try again",guess)

        elif guess >secret_number:
            print("too high try again",guess)

        else:
            print("congratulations you are successful guess correct  number",guess)
            print("The number of attempts in guessing the correct number is : ",attempts)
```

```python
number_guess_game()
```

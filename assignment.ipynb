Q 5) write a python program to computer the simple interest, p = 100000, r = 8%, t = 1 year.

```python
principal = 100000
rate = 8 # as percentage
time = 1 # in years

# Simple Interest formula: (P * R * T) / 100
simple_interest = (principal * rate * time) / 100

print(f"Principal (P): {principal}")
print(f"Rate (R): {rate}%")
print(f"Time (T): {time} year")
print(f"The Simple Interest is: {simple_interest}")
```

Q 7) write a python program to calculate the hypotenuse of a right angled triangle.

```python
import math

# Get input from the user for the two legs
try:
    side_a = float(input("Enter the length of the first leg: "))
    side_b = float(input("Enter the length of the second leg: "))

    if side_a <= 0 or side_b <= 0:
        print("Error: Both leg lengths must be positive numbers.")
    else:
        # Pythagorean theorem: hypotenuse = sqrt(a^2 + b^2)
        hypotenuse = math.sqrt(side_a**2 + side_b**2)

        print(f"\nLength of the first leg: {side_a:.2f}")
        print(f"Length of the second leg: {side_b:.2f}")
        print(f"The hypotenuse of the right-angled triangle is: {hypotenuse:.2f}")
except ValueError:
    print("Error: Invalid input. Please enter numerical values for the leg lengths.")
```

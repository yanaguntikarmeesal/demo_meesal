```python
# Q 1) enter the value from the user and check is it greater than 10 or 20 and not ?
a = int(input("enter the value : "))

if a > 10 :
  print("the number is greater than 10")
  if a > 20 :
      print("the number is greater than 20")
  else:
      print("the number is not greater  20")
else:
  print("the number is less than 10")
```
